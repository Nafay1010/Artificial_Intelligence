```python
import numpy as np
from copy import deepcopy
```

```python
class Node:
    def __init__(self, data, parent=None):
        self.left = None
        self.right = None
        self.down = None
        self.up = None
        self.data = data
        self.parent = parent

    def __str__(self):
        return str(self.data)

class Tree:
    def __init__(self, data):
        self.state = Node(data)
```
